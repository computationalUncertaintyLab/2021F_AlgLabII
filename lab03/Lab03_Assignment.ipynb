Lets suppose that the incident weekly cases in Northampton county can be modeled with an exponential distribution.
The exponential distribution is a continuous probability distribution characterized by the following probability density function

\begin{align}
    f(x|\mu) = \mu e^{-\mu x}
\end{align}

The exponential distribution has a single parameter $\mu$ and takes values in the interval from 0 to infinity.
