Lets suppose that the incident weekly cases in Northhampton county can be modeled with an exponential distribution.
The exponential distribution is a continuous probability distribution chaarcterized by the following pdf

\begin{align}
    p(x|\mu) = \mu e^{-\mu x}
\end{align}

The exponential distribution has a single parameter $\mu$.

### Q01: Please write down the loglikelihood for a single data point

\begin{align}
   g(\mu) =  p(x|\mu) = \log(\mu) -\mu x
\end{align}

### Q02: Please write down the loglikelihood for a $N$ data points assuming the data is i.i.d.

\begin{align}
    \log \left [p(x_{1},x_{2},\cdots,x_{N}|\mu) \right ] &= \sum_{i=1}^{N} \log(\mu) -\mu x_{i} \\ 
    &= N \log(\mu) - \mu \sum_{i=1}^{N} x_{i}
\end{align}

### Q03: Please write a function called loglikelihood that inputs a vector of data we assume is generated from an exp distribution and a parameter value $\mu$, and returns the loglikelihood from Q02. 

```r
loglikelihood = function(data,mu){
    N = length(data)
    return(N*log(mu) - mu*sum(data))
}
```

### Q04: Plot mu by the loglikelihood for the seq(0.001,0.010,0.0001)

```r
mus = seq(0.001,0.010,0.0001)
plot(mus,loglikelihood(newCases,mus))
```

### Q05: Write a wrapper function called LL that inputs a parameter mu and returns the loglikeihood, assuming the vector newCases is given as data

```r
LL = function(mu){
    return(loglikelihood(newCases,mu))
}
```

### Q06: Use the optimize function to find the MLE, assuming the interval of optimization is between 0 and 1

```r
O = optimize(LL,interval = c(0,1),maximum=TRUE)
O
```
