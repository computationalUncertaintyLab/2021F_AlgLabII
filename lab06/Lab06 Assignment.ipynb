## Q4 Is poverty related to BMI?

```r

plot(NHANES$BMI, NHANES$Poverty)
summary(lm( Poverty ~ BMI , data=NHANES))
summary(lm( BMI ~ Poverty , data=NHANES))
```

## Q5 Is poverty related to BMI controlling for age?

```r
summary(lm( BMI ~ Poverty + Age , data=NHANES))
```
