## Q4 Is poverty related to BMI?

$$y = 2 + 3*x + \epsilon$$

$$y = 2 + 3*(x+1) + \epsilon$$
$$y = 2 + 3*x+ 3 + \epsilon$$
$$\epsilon \sim N(0,\sigma^2)$$

```r
#hist(NHANES$Poverty)
#plot(NHANES$Poverty, NHANES$BMI)
fit <- lm(BMI ~ Poverty,data=NHANES)
hist(residuals(fit))

plot(NHANES$Poverty, NHANES$BMI)
abline(a=fit$coefficients[1] ,b=fit$coefficients[2],col='red')
```
