# Regression ( the workhorse of statistics)

Regression answers the question of how a particular outcome (dependent) variable relates to a particular explanatory (independent) variable. We use regression when we are interested in how one variable "explains" another.

Linear regression (as the name implies) assumes a linear relationship between the outcome and the explanatory variable. A line is defined by a slope and an intercept.

Now that we have some x data, lets relate the covariate data to the outcome through a straight line. 


$$Y_i \sim Normal(\beta_0 + \beta_1 X_i,\sigma^2)$$

### Checkpoint 


Set $\beta_0 = 2$ and $\beta_1  = 3$ to simulate the outcome variable. 

Suppose we don't know $\beta_0$ and $\beta_1$, we can estimate them from a set of observed data to find the relationship between an observed outcome and observed X. 

To do this we use the linear regression command in R. We can run a linear regression model using the `lm` command in R. The `lm` syntax is as follows

$$\text{lm(outcome ~ explanatory)}$$

# Application to COVID-19


Let's first download new cases and new deaths.

```r
data = read.csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv")
cases = colSums(data[which(data$Province_State == "Pennsylvania"), 100:200])

cases = as.numeric(cases)

cases = diff(cases,7)
plot(cases)



```

```r
data = read.csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv")
deaths = colSums(data[which(data$Province_State == "Pennsylvania"), 100:200])

deaths = as.numeric(deaths)

deaths = diff(deaths,7)
plot(deaths)



```

We can run a linear regression model using the `lm` command in R. The `lm` syntax is as follows

$$\text{lm(outcome ~ explanatory)}$$

```r
plot(cases,deaths)
```

What does the plot of cases against deaths tell us about the relationship?


```r
lm(deaths~ cases)
```

We refer to predicted values of our response variable as $\hat{Y}$ or "y-hat". This is the y-value we would expect if our straight line assumption were true.

# Residuals


We define a residual as the difference between the predicted value and the observed value of $Y$ for a given value of $X$. Mathematically, 

$$(\hat{Y}-Y)$$
