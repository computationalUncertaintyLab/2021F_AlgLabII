## Categorical Variables

Categorical variables are a little more complicated. Previously we just converted them to numeric values. For a first pass, this is fine. However, if we want to estimate a different effect size for each level of the category we need to include a coefficient for each value of the variable. 

```r
plot(NHANES$Education)
```

```r
lm(NHANES$Weight ~ NHANES$Education)
```

```r
lm(NHANES$Weight ~ as.numeric(NHANES$Education))
```

    Difference between the two is ability to have different effect of educaion levels by which education level you start at.

## What happens when our outcome is binary?

```r
lm( NHANES$SmokeNow ~NHANES$Weight )
```

Now we see the error we have previously talked about. Here we can convert the outcome to numeric.



```r
NHANES$SmokeNow
as.numeric(NHANES$SmokeNow)
```

```r
fit <- lm( as.numeric(NHANES$SmokeNow) ~NHANES$Weight )
```

Let's take a look at these residuals.

```r
fit <- lm( as.numeric(NHANES$SmokeNow) ~NHANES$Weight )
min(predict(fit))
```

# Logistic Regression 

Logistic regression to the rescue! Logistic regression is able to predict probabilities and keep everything between [0,1].

```r



fit <- glm(NHANES$SmokeNow ~NHANES$Weight ,family=binomial)
summary(fit)
```

### Use logistic regression with binary or categorical outcomes

```r
fit <- glm(NHANES$Education ~NHANES$Weight ,family=binomial)
summary(fit)
```
