# Multivariate Regression Interpretation


$$ y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \epsilon$$
$$ \epsilon \sim N(0,\sigma^2)$$

We know how to interpret $\beta_1$ in a univariate regression. It is the effect of a unit change of x on y. In multivariate regression we can no longer say this. 

We now say that $\beta_1$ is the effect of $x_1$ on $y$ at all levels of $x_2$. No matter what value we plug into $x_2$ we see the same effect on $y$ of $x_1$.

# NHANES Multivariate

Let's take a look at a real data example using NHANES.



```r
library(NHANES)

summary(NHANES$SmokeNow)
summary(NHANES$AlcoholDay)


fit <- lm( BPDiaAve ~  AlcoholDay +SmokeNow ,NHANES)
summary(fit)
```

We can interpret $\beta_1$ as the effect of drinking on blood pressure for all levels of smoking (yes/no).

```r
plot(fit)
```
